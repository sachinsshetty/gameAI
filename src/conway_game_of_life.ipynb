#### Conways Game of Life


```python
import numpy as np


def main():
    
    print("\n======================")
    print("Conway's Game of Life")
    print("======================\n")

    print("Input Selection for Initial Automaton \n")
    print("1 = Blinker \n2 = Glider \n3 = Gosper's Glider Gun \n4 = R-pentamino \n5 = Cross (Class 3)\n")

    choice = int(input("~ Choice: "))
    max_iter = int(input("~ Numbero of steps: "))
    print()
    
    file= open("ex_2_conway_out.txt","a+")
    
    
    # Configurations
    torus_width = 2
    # Game size
    max_y = 101
    max_x = 82
   
    matrix = get_pattern(choice, max_x, max_y, torus_width)
    
    for current_iter in range(1, max_iter+1):
        total_alive = np.count_nonzero(matrix)
        print("Step: ", current_iter, " - Living cells: " ,total_alive)
        file.write("%d\n" %(total_alive))
        print_state(matrix,torus_width)
        
        # Update step
        matrix = update_step(matrix, torus_width)

    file.close()
```

```python
def  get_pattern(choice, max_x, max_y, torus_width):
    

    # Centers
    c_y = max_y // 2
    c_x = max_x // 2
    matrix = np.zeros((max_y, max_x))
    
    if (choice ==1 ):
        print("Starting Blinker\n")
        matrix[4,3]=1
        matrix[4,4]=1
        matrix[4,5]=1

    elif (choice ==2 ):
        print("Starting Glider\n")
        matrix[5,3]=1
        matrix[5,4]=1
        matrix[5,5]=1
        matrix[4,5]=1
        matrix[3,4]=1
    elif (choice ==3 ):
        print("Starting Gosper's Glider Gun")
        # Left Block
        matrix[6,3]=1
        matrix[6,4]=1
        matrix[7,3]=1
        matrix[7,4]=1
        # Left structure
        matrix[6,13]=1
        matrix[7,13]=1
        matrix[8,13]=1
        matrix[5,14]=1
        matrix[9,14]=1
        matrix[4,15]=1
        matrix[4,16]=1
        matrix[10,15]=1
        matrix[10,16]=1
        matrix[7,17]=1
        matrix[5,18]=1
        matrix[9,18]=1
        matrix[6,19]=1
        matrix[7,19]=1
        matrix[8,19]=1
        matrix[7,20]=1
        # Right s ntructure
        matrix[4,23]=1
        matrix[5,23]=1
        matrix[6,23]=1
        matrix[4,24]=1
        matrix[5,24]=1
        matrix[6,24]=1
        matrix[2,27]=1
        matrix[3,27]=1
        matrix[3,25]=1
        matrix[7,25]=1
        matrix[7,27]=1
        matrix[8,27]=1
        # Right Block
        matrix[4,37]=1
        matrix[5,37]=1
        matrix[4,38]=1
        matrix[5,38]=1
        
    elif (choice ==4 ):
        #print("Starting r-pentamino")
        matrix[c_y,c_x+2]=1
        matrix[c_y,c_x+3]=1
        matrix[c_y+1,c_x+1]=1
        matrix[c_y+1,c_x+2]=1
        matrix[c_y+2,c_x+2]=1
    else :
        #print("Starting Cross (Class 3)")
        # Top side
        matrix[2,4]=1
        matrix[2,5]=1
        matrix[2,6]=1
        matrix[2,7]=1
        matrix[3,4]=1
        matrix[3,7]=1
        matrix[4,4]=1
        matrix[4,7]=1
        # Bottom side
        matrix[9,4]=1
        matrix[9,5]=1
        matrix[9,6]=1
        matrix[9,7]=1
        matrix[8,4]=1
        matrix[8,7]=1
        matrix[7,4]=1
        matrix[7,7]=1
        # Left side
        matrix[4,2]=1
        matrix[5,2]=1
        matrix[6,2]=1
        matrix[7,2]=1
        matrix[4,3]=1
        matrix[7,3]=1
        # Left side
        matrix[4,9]=1
        matrix[5,9]=1
        matrix[6,9]=1
        matrix[7,9]=1
        matrix[4,8]=1
        matrix[7,8]=1
    
    return matrix
```

```python
def get_neighbor_count(row_index, col_index, matrix, torus_width):

    # Top left
    tl_x = row_index -1
    tl_y = col_index -1
    # Top
    top_x = row_index -1
    top_y = col_index
    # Top right
    tr_x = row_index -1
    tr_y = col_index +1 
    # Left
    left_x = row_index
    left_y = col_index -1
    # Right
    right_x = row_index
    right_y = col_index + 1
    # Down left
    dl_x = row_index + 1
    dl_y = col_index -1
    # Down
    down_x = row_index + 1
    down_y = col_index
    # Down right
    dr_x = row_index + 1
    dr_y = col_index + 1

    max_x = matrix.shape[0] - torus_width
    max_y = matrix.shape[1] - torus_width

    min_x = torus_width
    min_y = torus_width

    count = 0

    if(tl_x >= min_x and tl_y >= min_y ):
        count = count + matrix[tl_x,tl_y]
        #print(count, "tl\n")
    if(top_x >= min_x):
        count = count + matrix[top_x,top_y]
        #print(count, "top\n")
    if(tr_x >= min_x and tr_y < max_y ):
        count = count + matrix[tr_x,tr_y]
        #print(count, "tr\n")
    if(left_y >= min_y):
        count = count + matrix[left_x,left_y]
        #print(count, "left\n")
    if(right_y < max_y):
        count = count + matrix[right_x,right_y]
        #print(count, "right\n")
    if(dl_x < max_x and dl_y >= min_y):
        count = count + matrix[dl_x,dl_y]
        #print(count, "dl\n")
    if(down_x < max_y):
        count = count + matrix[down_x,down_y]
        #print(count, "down\n")
    if(dr_x < max_x and dr_y < max_y):
        count = count + matrix[dr_x,dr_y]
        #print(count, "dr\n")

    return count

```

```python
def update_step(matrix,torus_width):
    
    rows,cols = matrix.shape
    
    
    row_range = range(torus_width, (rows-torus_width))
    
    col_range = range(torus_width, (cols-torus_width))
    
    new_matrix = np.zeros(matrix.shape)
    for row_index in row_range:
        for col_index in col_range:
            cell_value = matrix[row_index, col_index]
            count = 0
            count = get_neighbor_count(row_index, col_index, matrix, torus_width)
            if(cell_value == 1):
                # populated
                if(count == 1):
                    new_matrix[row_index,col_index] = 0  #solitary - death
                elif(count > 3):
                    new_matrix[row_index,col_index] = 0  #over population - death
                elif(count ==2 or count ==3):
                    new_matrix[row_index,col_index] = 1
                
                # for 2/3, do no modify value : lives
            else:
                # unpopulated
                if(count == 3):
                    new_matrix[row_index,col_index] = 1
            

    return new_matrix
    
    
```

```python
def print_state(matrix,torus_width):
    rows,cols = matrix.shape
      
    for row_ind in range(rows):

        for col_ind in range(cols):
            if(row_ind < torus_width):
                print("-",end=" ")
            elif(row_ind >= rows-torus_width):
                print("-",end=" ")
            elif(col_ind < torus_width):
                print("|",end=" ")
            elif(col_ind >= cols-torus_width):
                print("|",end=" ")
            else:
                if(matrix[row_ind, col_ind]==1):
                    print("X", end =" ")
                else:
                    print(".", end =" ")
        print("")
```

```python
if __name__=="__main__":
    main()
```

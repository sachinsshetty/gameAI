#### Conways Game of Life


```python
import numpy as np


def main():
    
    print("Implementation of Conway's Game of Life \n")
    print("Input Selection for Initial Automaton \n")
    print("Enter: 1 - Blinker, 2 - Glider, 3 - Gosper's Glider Gun, 4 - r-pentamino, 5 - Pattern(Class 3)")

    choice = int(input())
    
    matrix = get_pattern(choice)
    
    print("Enter max steps")
    max_iter = int(input())
    
    print("Initial Configuration")
    print_state(matrix)
    
    file= open("ex_2_conway_out.txt","a+")
    
    total_alive = np.count_nonzero(matrix)
    print("Steps : Initial ,  Alive cells : " ,total_alive)
    
    for current_iter in range(max_iter):
        new_matrix = update_step(matrix)
        
        total_alive = np.count_nonzero(new_matrix)
        print("Steps : ", current_iter, " Alive cells : " ,total_alive)
        file.write("Steps : %d \t Alive Cells %d \n " %(current_iter ,total_alive))
        print_state(new_matrix)
        matrix = new_matrix

    file.close()
```

```python
def print_state(matrix):
    rows,cols = matrix.shape
    
    for row_ind in range(rows):
        for col_ind in range(cols):
            if(matrix[row_ind, col_ind]==1):
                print("X", end =" ")
            else:
                print(".", end =" ")
        print("")
```

```python
def  get_pattern(choice):
    
    matrix = np.zeros( (45, 45) )

    
    if (choice ==1 ):
        print("Starting Blinker\n")

        #blinker
        matrix[2,1]=1
        matrix[2,2]=1
        matrix[2,3]=1

    elif (choice ==2 ):
        print("Starting Glider\n")
        
        #glider
        matrix[6,1]=1
        matrix[6,2]=1
        matrix[6,3]=1
        matrix[7,3]=1
        matrix[8,2]=1
    elif (choice ==3 ):
        print("Starting Gosper's Glider Gun")
        matrix[6,2]=1
        matrix[6,3]=1
        matrix[7,2]=1
        matrix[7,3]=1
        matrix[6,12]=1
        matrix[7,12]=1
        matrix[8,12]=1
        matrix[5,13]=1
        matrix[9,13]=1
        matrix[4,14]=1
        matrix[4,15]=1
        matrix[10,14]=1
        matrix[10,15]=1
        matrix[5,16]=1
        matrix[9,16]=1
        matrix[7,17]=1
        matrix[5,18]=1
        matrix[9,18]=1
        matrix[6,19]=1
        matrix[7,19]=1
        matrix[8,19]=1
        matrix[7,20]=1
        matrix[4,23]=1
        matrix[5,23]=1
        matrix[6,23]=1
        matrix[4,24]=1
        matrix[5,24]=1
        matrix[6,24]=1
        matrix[2,26]=1
        matrix[3,26]=1
        matrix[8,26]=1
        matrix[9,26]=1
        matrix[4,36]=1
        matrix[5,36]=1
        matrix[4,37]=1
        matrix[5,37]=1
        
    elif (choice ==4 ):
        
        #glider
        matrix[1,3]=1
        matrix[1,4]=1
        matrix[2,2]=1
        matrix[2,3]=1
        matrix[3,3]=1
        #r-pentamino
        print("Starting r-pentamino")
    else :
        print("Starting Pattern(Class 3)")
    
    
    # Initialize empty matrix as 0

    #matrix = np.zeros( (101, 82) )
    
    
    return matrix
```

```python
    
def get_neighbor_count(row_index, col_index, matrix):
    
    tl_x = row_index -1
    tl_y = col_index -1
    
    top_x = row_index -1
    top_y = col_index
    
    tr_x = row_index -1
    tr_y = col_index +1 
    
    left_x = row_index
    left_y = col_index -1
    
    right_x = row_index
    right_y = col_index + 1
    
    dl_x = row_index + 1
    dl_y = col_index -1
    
    down_x = row_index + 1
    down_y = col_index
    
    dr_x = row_index + 1
    dr_y = col_index + 1
    
    max_x = matrix.shape[0]
    max_y = matrix.shape[1]
    count = 0
    
    if(tl_x >= 0 and tl_y >= 0 ):
        count = count + matrix[tl_x,tl_y]
        #print(count, "tl\n")
    if(top_x >= 0):
        count = count + matrix[top_x,top_y]
        #print(count, "top\n")
    if(tr_x >= 0 and tr_y < max_y ):
        count = count + matrix[tr_x,tr_y]
        #print(count, "tr\n")
    if(left_y >= 0):
        count = count + matrix[left_x,left_y]
        #print(count, "left\n")
    if(right_y < max_y):
        count = count + matrix[right_x,right_y]
        #print(count, "right\n")
    if(dl_x < max_x and dl_y >= 0):
        count = count + matrix[dl_x,dl_y]
        #print(count, "dl\n")
    if(down_x < max_y):
        count = count + matrix[down_x,down_y]
        #print(count, "down\n")
    if(dr_x < max_x and dr_y < max_y):
        count = count + matrix[dr_x,dr_y]
        #print(count, "dr\n")
        
    return count
    
    
```

```python
def update_step(matrix):
    
    rows,cols = matrix.shape
    
    new_matrix = np.zeros(matrix.shape)
    for row_index in range(rows):
        for col_index in range(cols):
            cell_value = matrix[row_index, col_index]
            count = 0
            count = get_neighbor_count(row_index, col_index, matrix)
            if(cell_value == 1):
                # populated
                if(count == 1):
                    new_matrix[row_index,col_index] = 0  #solitary - death
                elif(count > 3):
                    new_matrix[row_index,col_index] = 0  #over population - death
                elif(count ==2 or count ==3):
                    new_matrix[row_index,col_index] = 1
                
                # for 2/3, do no modify value : lives
            else:
                # unpopulated
                if(count == 3):
                    new_matrix[row_index,col_index] = 1
            
    
    return new_matrix
    
    
```

```python
if __name__=="__main__":
    main()
```
